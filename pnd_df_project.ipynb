```python
import random
```

```python
"""
adding new column Country_population by assigning a random number to each row in the df for that column
df['Country_population'] = : creates a new column "Country_population"
df.apply( : applies the function to the DataFrame
lambda x: random.randint(1000000, 1000000000) : the function being applied to each row. This is an anonymous function (also known as a lambda function) that takes a single argument x (representing each row of the DataFrame) and returns a random integer between 1000000 and 1000000000.
, axis=1) : specifies that the function should be applied to each row (as opposed to each column)
"""
datatable['Country_population'] = datatable.apply(lambda x: random.randint(1000000, 1000000000), axis=1)
```

```python
datatable.head()
```

```python
datatable['Companies'] = datatable.apply(lambda x: random.randint(100, 10000), axis=1)
datatable.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#show plots in notebook
%matplotlib inline
```

```python
# plot that groups count of countries with a specific range of number of companies, bins is the number of blue bins on the plot 
visual_1 = sns.displot(datatable['Companies'], bins=15)
# ignoring warnings just in case
import warnings
warnings.filterwarnings('ignore')
```

```python
# lmplot shows the correlation between Country_population and number of Companies per Currency
visual_2 = sns.lmplot(data=datatable, x='Country_population', y='Companies', fit_reg=False, hue='Currency')
```
